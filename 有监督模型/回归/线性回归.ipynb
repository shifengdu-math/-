```python
import numpy as np
import pandas as pd
import seaborn as sn
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import MinMaxScaler   # 数据归一化
from sklearn.model_selection import train_test_split  # 划分训练集和测试集
import matplotlib.pyplot as plt
```

```python
housing = pd.read_csv(fr'C:\Users\shifeng.du\Desktop\github\sklearn-model\data\Linear Regression\kc_train.csv', header = None)
housing.columns = ['date','price','bedroom','bath','room_area','car_area','floor','score','floor_area','basement_area','construction_year',
                   'restore_year','latitude','longitude']
housing
y = housing.loc[:,['price']]
x = housing.drop(columns='price')
```

```python
'''特征缩放----归一化'''



'''这里测试集应该划分出验证集和测试集。。。。。'''
minmax_scaler = MinMaxScaler()
minmax_scaler.fit(x)
scaler_housing = minmax_scaler.transform(x)
scaler_housing = pd.DataFrame(scaler_housing, columns=x.columns)
scaler_housing
```

```python
x_train, x_verify, y_train, y_verify = train_test_split(scaler_housing, y, test_size=0.2, random_state=42)
x_train
```

```python
LR_reg = LinearRegression()
LR_reg.fit(x_train, y_train)

preds = LR_reg.predict(x_verify)
mse = mean_squared_error(preds, y_verify)

print(mse)
# plt.figure(figsize = (10,7))
# num = 200
# x = np.arange(1,num + 1)
# plt.plot(x, y_train[:num], label = 'target')
# plt.plot(x, preds[:num], label = 'preds')
# plt.legend(loc = 'upper left')
# plt.show()
```

```python
'''预测'''
test = pd.read_csv(fr'C:\Users\shifeng.du\Desktop\github\sklearn-model\data\Linear Regression\kc_test.csv',header = None)
test.columns = x_train.columns

minmax_scaler_test = MinMaxScaler()
minmax_scaler_test.fit(test)
scaler_test = minmax_scaler_test.transform(test)
scaler_test = pd.DataFrame(scaler_test, columns=test.columns)

result = LR_reg.predict(scaler_test)
test_result = pd.concat([test, pd.DataFrame(result).rename(columns={0:'price'})], axis= 1)
test_result
```
